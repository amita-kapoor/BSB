## Activity 4.1: Creating and Viewing a Simple Physics Dataset

We will begin with a small synthetic dataset based on Newton's second law. When you apply a greater force on an object, its acceleration increases in a roughly linear way. We will create a tiny dataset with two columns: Force and Acceleration.

```python
import pandas as pd

data = {
    "Force": [2, 4, 6, 8, 10, 12],
    "Acceleration": [1.1, 2.0, 2.9, 4.1, 5.0, 6.1]
}

df = pd.DataFrame(data)
df
```

Each row shows how much acceleration we get when a certain force is applied. The values are simple and slightly imperfect, just like real measurements.

## Activity 4.2: Visualising the Data with a Scatter Plot

Looking at the table is helpful, but seeing the pattern visually is even more powerful. A scatter plot shows each pair of values as a dot on a graph. This helps you notice whether the points form a line, a curve, or no pattern at all.

> ðŸ“˜ **Learning Note**
>
> A **scatter plot** shows each pair of values as a single dot on a graph.
> One value is placed on the horizontal axis and the other on the vertical axis.
>
> Scatter plots are useful when we want to study the **relationship between two quantities**,
> such as force and acceleration.
>
> There are other kinds of plots too, like line plots, bar charts, and histograms.
> We are using a scatter plot here because each measurement is a separate observation,
> and we want to see whether the points follow a straight pattern.
>
> If the points roughly form a straight line, it suggests a linear relationship,
> which we can model using a line of best fit.


```python
import matplotlib.pyplot as plt

plt.scatter(df["Force"], df["Acceleration"])
plt.xlabel("Force (N)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Force vs Acceleration")
plt.grid(True)
plt.show()
```

You will see that the dots rise as the force increases. They form a roughly straight pattern. This tells you that a straight line would be a good model to summarise this relationship.

## Activity 4.3: Fitting a Line of Best Fit

A line of best fit is a line that follows the general trend of the data. It does not pass through every point, but it stays close to all of them. The line is described by the familiar equation:

$$y = m * x + c$$

where $m$ is the slope and $c$ is the intercept.

You can calculate $m$ and $c$ using numpy, another python library we used in Unit 1.

```python
import numpy as np

x = df["Force"]
y = df["Acceleration"]

m, c = np.polyfit(x, y, 1)
m, c
```

> **_Note_**
>The function `np.polyfit` is a tool from the numpy library that helps Python find a mathematical pattern in a set of data points. In our case, we asked it to find a straight line that matches the Force vs Acceleration pattern as closely as possible.
> The three pieces in the command have simple meanings:
> * The first part, `x`, is the list of values along the horizontal axis.
> * The second part, `y`, is the list of values along the vertical axis.
> * The third part, `1`, tells numpy that we want a line. A line is also called a **first degree polynomial**, which is why we use the number 1 here.
> The function then returns two numbers. The first is the slope $m$, and the second is the intercept $c$. These two numbers together describe the best possible straight line for the data.

Although you are using it here to study relationship between force and acceleration, this function can be used in many other situations that for instance:

* A pattern where cost increases steadily with quantity
* A pattern where distance increases at a constant speed
* A pattern where marks increase in a regular way with hours of study
* A pattern where temperature rises steadily over time in a classroom experiment

Any time the data seems to follow a straight line, `np.polyfit` helps you find the equation of that line. It is like having a small mathematical assistant who looks at all the points and chooses the line that fits best.

You will use this idea again in future units when you work with patterns and simple models.

Now let us plot the line along with the data points.

```python
plt.scatter(x, y, label="Data Points")
plt.plot(x, m * x + c, color="red", label="Line of Best Fit")
plt.xlabel("Force (N)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Line of Best Fit for Force vs Acceleration")
plt.legend()
plt.grid(True)
plt.show()
```

You now have a mathematical model that describes the relationship. The slope tells you how fast acceleration increases when force increases. The intercept tells you where the line begins.

## Activity 4.4: Making a Prediction

Once you have the line, making a prediction is simple. Choose a new value for force and plug it into the equation. For example, what acceleration might we expect if the force is 5.5 newtons?

```python
new_force = 5.5
predicted_acc = m * new_force + c
predicted_acc
```

This predicted value does not come from the table. It comes from the pattern in the data. You have used mathematics to estimate something new based on what you already know.

Try another value if you like.

```python
new_force = 9
m * new_force + c
```

This is the basic idea behind many predictive models in science, artificial intelligence and data analysis.
