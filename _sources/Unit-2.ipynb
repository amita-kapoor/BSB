A DataFrame is simply a digital table, like a table you draw in your notebook, with:
> * column names at the top

## Activity 2.2: Becoming a Statistical Detective

Now that your dataset is ready, you can begin examining it just like a detective examines clues. One of the first ways to understand a list of numbers is to look at what is ‚Äútypical‚Äù or ‚Äúcentral.‚Äù You have already learned three useful ideas: the **mean**, the **median**, and the **mode**. Python can calculate these for you using simple commands.

We will begin with the **Temperature** column. The DataFrame stores this column under the name "Temperature", so we can ask Python to look at it and compute these values.

```python
mean_temp = df["Temperature"].mean()
median_temp = df["Temperature"].median()
mode_temp = df["Temperature"].mode()[0]

print(f"Mean Temperature: {mean_temp} \nMedian Temperature: {median_temp} \nMode Temperature: {mode_temp}")
```

When you run this, Python will give you three numbers. These help you understand the overall behaviour of the temperature across the ten days. The mean tells you the average, the median tells you the middle value when the temperatures are arranged in order, and the mode tells you which temperature occurred most often.

###### Try doing the same exercise manually and compare the time it took you to calculate these numbers versus the computer.

You can do the same for the Humidity column. This lets you compare the two measures and get a sense of whether humidity varies more or less than temperature in this small dataset.

```python
mean_hum = df["Humidity"].mean()
median_hum = df["Humidity"].median()
mode_hum = df["Humidity"].mode()[0]

print(f"Mean Humidity: {mean_hum} \nMedian Humidity: {median_hum} \nMode Humidity: {mode_hum}")
```

As you look at the results, notice how these numbers act like summaries of the whole column. Instead of examining all ten values one by one, these three measures help you understand the overall pattern at a glance. This is exactly what a Data Detective does‚Äîuse a few smart calculations to make sense of many values.

If you want to compare columns directly, you can print the results together:

```python
print("Temperature:", mean_temp, median_temp, mode_temp)
print("Humidity:   ", mean_hum, median_hum, mode_hum)
```

## Activity 2.3: Visualizing Humidity with a Histogram

Numbers become much easier to understand when you can see them. In the same way that ancient scholars watched cloud shapes or star positions to notice patterns, you can look at a picture of your data to understand how values are spread out. One of the simplest and most powerful ways to do this is with a histogram.

A histogram shows how often certain ranges of values appear. If you think of the humidity values as grains collected in different pots, a histogram tells you how full each pot is. Higher bars mean those values occurred more frequently; lower bars mean they were less common.

Python can draw this picture for you using the plotting library matplotlib. We will create a histogram of the Humidity column.

```python
import matplotlib.pyplot as plt

plt.hist(df["Humidity"], bins=5, edgecolor="black")
plt.xlabel("Humidity (%)")
plt.ylabel("Frequency")
plt.title("Humidity Distribution")
plt.show()
```

```python
plt.hist(df["Humidity"], bins=10, edgecolor="black")
plt.show()
```

More bins divide the humidity values into smaller ranges, giving you a more detailed picture. Fewer bins give you a simpler overview. A Data Detective tries both to get a clear understanding of the data.

In the next activity, you will continue this detective work by interpreting the patterns you see and connecting them to the idea of making simple inferences from data.

## Activity 2.4: Interpreting the Patterns You See

Now that you have drawn a histogram, you can begin to interpret it the way a detective interprets clues. A histogram is more than a picture; it is a summary of many values at once. Instead of reading every humidity number, you can look at the bars and understand what the data is trying to tell you.

Start by looking at which bars are the tallest. Taller bars mean that the humidity values in that range occurred more often. Shorter bars tell you that those values were less common. This simple observation already tells you something about the overall pattern of humidity for the ten days.

Let‚Äôs look at the actual values:

```python
df["Humidity"]
```

Now compare the list with the shape of your histogram. You will notice how the histogram gathers nearby values into a group and shows them as a single bar. This makes it easier to see whether the humidity values are mostly high, mostly low, or nicely spread out in the middle.

To help you reflect as a Data Detective, think about these simple questions:

* Are there any humidity values that seem unusual or far from the others?
* If you had to describe the humidity for these ten days to someone who did not see the table or the plot, what would you say?

You can also compare the histogram with the statistical measures you calculated earlier. For example, if the mean and median humidity are close together, and the histogram looks fairly balanced on both sides, that tells you the humidity values are not too skewed or uneven.

Sometimes it helps to print the mean and median again to keep them in mind while you look at the histogram:
