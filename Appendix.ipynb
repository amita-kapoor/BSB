## Create a Small Dataset in Python

Run the code below. It will create a table and save it as a file called weather_data.csv in your notebook folder.

```python
import pandas as pd

# Creating a small weather dataset with 10 rows
data = {
    "Temperature": [22, 24, 25, 23, 26, 27, 28, 24, 22, 23],
    "Humidity":    [60, 62, 58, 65, 70, 72, 68, 64, 59, 61],
    "AirPressure": [1012, 1010, 1011, 1013, 1014, 1012, 1011, 1013, 1010, 1012]
}

df = pd.DataFrame(data)

# Save the dataset as a CSV file
df.to_csv("weather_data.csv", index=False)
```
